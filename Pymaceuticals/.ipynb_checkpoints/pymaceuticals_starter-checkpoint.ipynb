## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merge_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")
# Display the data table for preview
merge_df.to_csv('test.csv')
merge_df
```

```python
# Checking the number of mice.
Mouse = [len(set(merge_df['Mouse ID']))]
total_mouse={'Total Mouses': Mouse}
pd.DataFrame(total_mouse)
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates = merge_df[merge_df.duplicated(subset=['Mouse ID', 'Timepoint'])]
a = duplicates['Mouse ID'].iloc[0]
a
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse = merge_df.loc[merge_df['Mouse ID'] == a]
duplicate_mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = merge_df.loc[merge_df['Mouse ID'] != a]
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
Mouse1 = [len(set(clean_df['Mouse ID']))]
total_mouse1 ={'Total Mouses': Mouse1}
pd.DataFrame(total_mouse1)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.

drug_group = clean_df.groupby(['Drug Regimen'])
mean = drug_group['Tumor Volume (mm3)'].mean()
median = drug_group['Tumor Volume (mm3)'].median()
variance = drug_group['Tumor Volume (mm3)'].var()
stdv = drug_group['Tumor Volume (mm3)'].std()
sem = drug_group['Tumor Volume (mm3)'].sem()
summary = pd.DataFrame({"Mean": mean, "Median": median,"Variance": variance, "Standard Deviation": stdv,"SEM":sem})
summary
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function
drug_group = clean_df.groupby(['Drug Regimen'])
mean = drug_group['Tumor Volume (mm3)'].mean()
median = drug_group['Tumor Volume (mm3)'].median()
variance = drug_group['Tumor Volume (mm3)'].var()
stdv = drug_group['Tumor Volume (mm3)'].std()
sem = drug_group['Tumor Volume (mm3)'].sem()
summary = pd.DataFrame({"Mean": mean, "Median": median,"Variance": variance, "Standard Deviation": stdv,"SEM":sem})
summary
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
Maximum_values = clean_df.loc[clean_df.groupby("Mouse ID")['Timepoint'].idxmax()]
drugs = ['Capomulin','Ramicane', 'Infubinol','Ceftamin']
filtered = []
for x in drugs:
    filtered.append(Maximum_values.loc[Maximum_values['Drug Regimen'] == x])
Maximum_values
```

```python

# Put treatments into a list for for loop (and later for plot labels)
# Locate the rows which contain mice on each drug and get the tumor volumes 
# add subset    
# Determine outliers using upper and lower bounds
drugs = ['Capomulin','Ramicane', 'Infubinol','Ceftamin']
full_data = {}
# Create empty list to fill with tumor vol data (for plotting)
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for x in drugs:
    med = Maximum_values.loc[Maximum_values['Drug Regimen'] == x]
    list1 = list(med['Tumor Volume (mm3)'])
    full_data[x]= list1    
    
    quartiles = med['Tumor Volume (mm3)'].quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    print(f'     ********* {x} *********')
    print(f"The lower quartile of {x} is: {round(lowerq,2)}")
    print(f"The upper quartile of {x}is: {round(upperq,2)}")
    print(f"The interquartile range of {x} is: {round(iqr,2)}")
    print(f"The the median of {x} is: {round(quartiles[0.5],2)} ")

    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f"Values below {round(lower_bound,2)} could be outliers.")
    print(f"Values above {round(upper_bound,2)} could be outliers.")
    below = med.loc[med['Tumor Volume (mm3)'] < lower_bound].count() 
    above =  med.loc[med['Tumor Volume (mm3)'] > upper_bound].count()
    outliers = above['Tumor Volume (mm3)'] + below['Tumor Volume (mm3)']
    print('\x1b[1;31m'+f"   {outliers} values could be outliers."+'\x1b[0m')
    print('-------------------------------------------------')
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
red_square = dict(markerfacecolor='r', marker='s')
w = pd.DataFrame(full_data)
w.boxplot(flierprops=red_square).set_title('Drug Regimen')
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
mouse_id = 'r554'
capomulin = clean_df.loc[clean_df['Drug Regimen'] == 'Capomulin']
mouse_data = capomulin.loc[capomulin['Mouse ID'] == mouse_id]
x_vals = mouse_data['Timepoint']
y_vals = mouse_data['Tumor Volume (mm3)']
plt.plot(x_vals,y_vals,'gs-', linewidth=2, markersize=12)
plt.title('Timepount vs. Tumor volume (Capomulin)')
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume (mm3)')
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
average = capomulin.groupby('Mouse ID').mean()
x_values = average['Weight (g)']
y_values = average['Tumor Volume (mm3)']
plt.scatter(x_values,y_values,marker="o",facecolors="green", edgecolors="black", alpha=0.60)
plt.title('Mouse weight vs. Average tumor volume (Capomulin)')
plt.xlabel('Weight (g)')
plt.ylabel('Average tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation = st.pearsonr(x_values,y_values)
print(f"The correlation between both factors is {round(correlation[0],2)}")
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values,marker="o",facecolors="green", edgecolors="black", alpha=0.60)
plt.title('Mouse weight vs. Average tumor volume (Capomulin)')
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,36),fontsize=15,color="blue")
plt.xlabel('Weight (g)')
plt.ylabel('Average tumor Volume (mm3)')
plt.show()
```
